```python
# doing some imports here
import numpy as np
import scipy as sp
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# enable multiple output in a cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# see some basic properties about datas
data_1 = pd.read_csv(r'C:\Users\riema\un-comtrade\dataset\comtrade.csv')
# example_Data = pd.read_csv(r'C:\Users\riema\un-comtrade\dataset\comtrade.csv', skiprows=1)
data_1[['Period']]
cols = data_1.columns
dims = data_1.shape
vals = data_1.values
vals
```

```python
# visualize the relation between trade values and country each month
country = data_1['Trade Value (US$)'].groupby([data_1['Reporter'], data_1['Period']]).sum().unstack()
country.plot(kind="bar", figsize=(25, 10))
```

```python
# visualize the relation between trade values and time for each country
# change the position the variables in comparison with the last cells
country = data_1['Trade Value (US$)'].groupby([data_1['Period'], data_1['Reporter']]).sum().unstack()
country.plot(kind="bar", figsize=(25, 10))
```

```python
# visualize the relation between trade values and trade flow for each reporter
# to see which kind of trade flow can make the most profit?
country = data_1['Trade Value (US$)'].groupby([data_1['Trade Flow'], data_1['Reporter']]).sum().unstack()
country.plot(kind="bar", figsize=(25, 10))
```

```python
# more or less like summary information compared to the last cell
fig_2, ax_2 = plt.subplots(2,1)

data_2 = data_1.groupby('Trade Flow')['Trade Value (US$)'].sum()
data_2.plot(kind='barh', ax=ax_2[0], color='r', alpha=0.5)

data_3 = data_1.groupby('Reporter')['Trade Value (US$)'].sum()
data_3.plot(kind='bar', ax=ax_2[1], color='k', alpha=0.5, figsize=(23, 10))
```

```python
# counting the frequency of reporters in the recorded files
# notation: there are 2 reporters' name missing in this file, but the trade value remains there
# have to deal with this embarrassing situation
data_keep_type = pd.read_csv(r'C:\Users\riema\un-comtrade\dataset\comtrade.csv', keep_default_na=False)
fig_1 = plt.figure()
ax_1 = fig_1.add_subplot(111)
fig_1.set_size_inches(20, 5)
num_reporter = len(data_keep_type['Reporter'].unique())
ax_1.hist(data_keep_type['Reporter'], bins=num_reporter, alpha=0.5)
plt.title('the number of trade flows')
plt.xlabel('Reporters')
plt.xticks(rotation=90) # the original angle seems not friendly
plt.ylabel('Trade Flows')

# use a line to show the average number of report among all reporters
mean = len(data_keep_type) / num_reporter
plt.axhline(mean, color='r')

plt.show()
```

```python
# Oh, the period are ordered alphabetically if I use "Period Desc." for the line???
period_count = data_1.groupby('Period').size().reset_index(name='Count')
period_count.plot(x = 'Period', y = 'Count', color = 'b', marker = 'o', legend = False)
plt.title('Trade Count By Period')
plt.xlabel('Period')
plt.ylabel('Period Count')
plt.show()
```

```python
trade_type = data_1.groupby(['Reporter', 'Trade Flow']).size().reset_index(name = 'Count')
trade_type_count = trade_type.groupby(['Reporter', 'Trade Flow']).Count.sum()
trade_type_count.unstack().plot(kind='bar', stacked=True, color=['r', 'b','g','y'], figsize=(15, 10))
plt.title('Trade Count By Type')
plt.ylabel('Trade Flow Count')
plt.tick_params(labelsize = 8)
plt.show()
```

```python
data_1['Trade Value (US$)'].min()
data_1['Trade Value (US$)'].max()
data_1['Trade Value (US$)'].mean()
data_1['Trade Value (US$)'].median()
```

```python
# set the boundaries somewhat randomly by just have a glance at the mean, median data etc
# can set the boundaries more precisely later
data_1['Trade Amount'] = pd.cut(data_1['Trade Value (US$)'], bins = [0, 2e10, 5e10, 3e12], labels=['Low', 'Normal', 'High'])
group_pd = data_1.groupby(['Period', 'Trade Amount']).size().unstack()
group_pd[['Low', 'Normal', 'High']].plot(kind = 'bar')
plt.title('Trade Count By Period and Amount')
plt.show()
```

```python
data_1['Trade Amount'].value_counts().plot(kind='pie', autopct='%.2f%%', startangle=90)
# plt.axis('equal')
plt.title('Trade Amount Pie')
plt.show()
```
